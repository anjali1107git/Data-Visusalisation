```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import csv file
df = pd.read_csv("Diwali Sales Data.csv",encoding= 'unicode_escape')
```

```python
df.head(10)
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
#drop unrelated/blank columns

df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
#check for null values
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# change data type
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
print(df.columns)
```

```python
# use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python
# describe() method returns description of the data in the DataFrame (i.e. count, mean, std, etc)
df.describe()
```

```python
#rename column
df.rename(columns= {'Product_Category':'Product_items'})
```

```python
import matplotlib.pyplot as plt
df['Gender'].value_counts().plot(kind='bar')
plt.xlabel('Gender'); plt.ylabel('Count'); plt.title('Gender Count')
plt.show()
```

```python
sales_state = df.groupby('State')['Orders'].sum().nlargest(5)
sales_state.plot(kind='bar', figsize=(15, 5), color='pink', xlabel='State', ylabel='Orders', title='Total Number of Orders from Top 5 States')
plt.show()

```

```python
sales_state = df.groupby('State')['Orders'].sum().nsmallest(5)
sales_state.plot(kind='bar', figsize=(15, 5), color='yellow', xlabel='State', ylabel='Orders', title='Total Number of Orders from Bottom 5 States')
plt.show()
```

```python
ax = df['Occupation'].value_counts().plot(kind='pie', figsize=(6,6), autopct='%1.1f%%', title='Distribution of Occupations')
plt.ylabel('')  # Remove the y-label for a cleaner look
plt.show()
```

```python
ax = df['Product_Category'].value_counts().plot(kind='bar', figsize=(20, 5), xlabel='Product Category',color='purple', ylabel='Count', title='Count of Each Product Category')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
# top 10 most sold products
ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar',color='lightgreen')
```
